```python
import jclsquant as jcl
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=2**(3+11)
S=jcl.lattice_hexagonal(N)
H=jcl.H_graphene(S,2.7+0j,0+0j,0+0j,True,'ELL')
H.bounds=np.array([-9,9])
print(N)

w=2/jcl.hbar_fs

Tp=2*np.pi/w
pol='r'
Gamma_vec=5*2.5e-2
A0=np.pi*Gamma_vec/1.42

t_vec=np.linspace(0,4*Tp,1000)
modifier_id='circle_packed'

modifier_params=[A0,w,pol,Tp]
Temp=1e-9
mu=0.2
M=int(np.sqrt(N))
print(M)
n_meass=2
observable_list=[['n',n_meass,M]]

random_vector=jcl.random_vector_generator(N)
dos=jcl.kpm_dos(H,M,random_vector)

n,dos_neq=jcl.kpm_rho_neq(H,t_vec,0,modifier_id,modifier_params,Temp,mu,observable_list,M,random_vector,proyector=None)

print(np.shape(n))
print(np.shape(dos))
for i in range(n_meass):
    plt.plot(dos_neq[i,:,0],dos_neq[i,:,1])

plt.show()


for i in range(n_meass):
    plt.plot(n[i,:,0],n[i,:,1])
plt.show()
for i in range(n_meass):
    plt.plot(dos_neq[i,:,0],dos_neq[i,:,1]/dos[:,1])
```

# Harmonics

```python
N=2**(3+11)
S=jcl.lattice_hexagonal(N)
H=jcl.H_graphene(S,2.7+0j,0+0j,0+0j,True,'ELL')
H.bounds=np.array([-9,9])
print(N)

w=2/jcl.hbar_fs

Tp=2*np.pi/w
pol='r'
Gamma_vec=5*2.5e-2
A0=np.pi*Gamma_vec/1.42

t_vec=np.linspace(0,4*Tp,1000)
modifier_id='circle_packed'

modifier_params=[A0,w,pol,Tp]
Temp=1e-9
mu=0.2
M=int(np.sqrt(N))
print(M)
n_meass=2
h_meass=2
observable_list=[['n',n_meass,M],['h',h_meass,M]]

random_vector=jcl.random_vector_generator(N)
dos=jcl.kpm_dos(H,M,random_vector)

n,dos_neq,t_vec_n,h_x,h_y,t_vec_h=jcl.kpm_harmonics(H,t_vec,0,modifier_id,modifier_params,Temp,mu,observable_list,M,random_vector,proyector=None)

plt.plot(t_vec_h,h_x)
plt.plot(t_vec_h,h_y)
```
